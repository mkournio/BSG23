```python
# CARGO LOS PAQUETES
import libwwz
import libwwz.wwz as wwz
import libwwz.plot_methods as wwz_plot
import os  
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import matplotlib as mpl
# cargo la libreria pandas que sirve para leer archivos con multiples columnas
import pandas as pd
```

```python
# cargo los datos del primer sector (sector 5)
#voy a usar la misma sentencia pero cambiando el sep=" " porque cambie de archivo al normalizado y funciona mejor asi, ya que esta equiespaciado
HD34085_5 = pd.read_csv("/home/aldana/Documentos/Paper_Michalis/HD34085_5_DMAG_FFI", sep=' ', header=None, names=['tiempo', 'mag' , "mag error"])

# Ahora voy a probar guardar cada valor de las columnas en variables y pedir en pantalla esas variables. Asi tambien me sirve para graficar en un futuro
tiempo = HD34085_5["tiempo"]
magnitud = HD34085_5["mag"]
error= HD34085_5["mag error"]
```

```python
# graficamos nuestra curva para chequear
plt.figure(figsize=(15,6))    
plt.plot(tiempo, magnitud,"-o") 
# Invertir el eje y
plt.gca().invert_yaxis()

plt.show()
```

```python
# De acuerdo con la Frecuencia de Nyquist (frecuencia  maxima) dado que esta curva tiene muy pocos datos deberia ser de fmax=360. Es demasiado para una resolucion de WWZ. Por lo tanto probaremos solo con fmax=10 a ver si encontramos algo
# En cuanto a la frecuencia minima, vamos a considerar  como fmin=2/Long serie(dias). 

#Calculamos ahora la extension de la curva de luz, para tener asi la frecuencia min
long = HD34085_5['tiempo'].max()-HD34085_5['tiempo'][0]
fmin=2/long
print(fmin)
```

```python
f = 2
w = 2 * np.pi * f        
c = 1/(2*w**2)

#damos los valores de frecuencia
freq_low = fmin
freq_high = 1
freq_steps = 0.001  
freq_lin = [freq_low, freq_high, freq_steps]

#la libreria libwwz trabaja con array np y no con dataframe, por eso los pasamos a numpy
tt = tiempo.to_numpy()     
dd = magnitud.to_numpy()
```

```python
#comenzamos con el analisis wavelet

WWZ_simple_linear = wwz.wwt(tt, dd, 100, freq_lin, c, 'linear')   # el numero 100 es la resolucin sobre el eje x 
# est utilizando la funcin "wwt" del paquete "wwz"

arr_transposed = np.transpose(WWZ_simple_linear[0]) #El cdigo est utilizando la funcin `transpose` de la librera NumPy de Python para transponer la matriz `WWZ_simple_linear[0]`. La transposicin de una matriz implica intercambiar sus filas por sus columnas, de modo que las filas se convierten en columnas y viceversa. El resultado se almacena en una nueva variable llamada `arr_transposed`.
indices_maximos = np.argmax(WWZ_simple_linear[2], axis=0) #La variable "WWZ_simple_linear" es un arreglo de tres dimensiones y se est buscando el ndice del valor mximo en la dimensin 2 del arreglo. La funcin "argmax" de NumPy devuelve el ndice del valor mximo a lo largo del eje especificado. El parmetro "axis=0" indica que se est buscando el valor mximo para cada columna de la dimensin 2.
```

```python
# Este seria el grafico ideal para mirar todo junto y usar de modelo para las publicaciones

# Importamos la biblioteca mpl_toolkits.axes_grid1
from mpl_toolkits.axes_grid1 import make_axes_locatable

#Creamos la figura con dos subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw={'width_ratios': [3, 1]})

# Dibujamos el escalograma en el primer subplot
heatmap = ax1.pcolormesh(WWZ_simple_linear[0], WWZ_simple_linear[1], WWZ_simple_linear[2], cmap='magma') #la funcion pcolormesh es para hacer el escalograma


# Agregamos la barra de colores al escalograma
divider = make_axes_locatable(ax1)
cax = divider.append_axes("left", size="5%", pad=0.2)
colorbar = fig.colorbar(heatmap, cax=cax)
colorbar.ax.yaxis.set_ticks_position('left')  # movemos las marcas de la barra de colores al lado izquierdo
colorbar.ax.tick_params(labelsize=8)  # reducimos el tamao de las marcas de la barra de colores

#Definimos las caracteristicas para el primer subplot
ax1.set_xlabel('Time')
#ax1.set_ylabel('Freq')   #le saco el nombre, porque lo lleva el grafico de potencia
ax1.set_yticklabels([])   #le saco los valores, porq tambien los lleva el grafico de potencia
ax1.set_yticks(np.arange(0.1, 1.1, 0.1))  # especificamos las marcas que deseamos mostrar
ax1.set_ylim([0.1, 1])  # establecemos el rango del eje y hasta 1

# Dibujamos el grfico de potencia promedio en el segundo subplot
df = pd.DataFrame(WWZ_simple_linear[2])
df2 = pd.DataFrame(WWZ_simple_linear[1])
df11 = df.mean()
df22 = df2.mean()
time1 = df22.to_numpy() 
Averag = df11.to_numpy() 

ax2.plot(Averag, time1, "-o", markersize=2, color='red')
ax2.set_xlabel("Potencia")
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.set_ylabel("Freq")
#ax2.invert_yaxis()
# Modificamos el eje y del segundo subplot
ax2.set_ylim([0.1, 1])  # establecemos el rango del eje y hasta 1


# Agregamos la lnea horizontal en el primer subplot
#ax1.axhline(y=0.21, color='black', linestyle='--')
#ax1.axhline(y=0.42, color='black', linestyle='--')
#ax1.axhline(y=0.89, color='black', linestyle='--')
# Agregamos la lnea horizontal en el segundo subplot
#ax2.axhline(y=0.21, color='black', linestyle='--')
#ax2.axhline(y=0.42, color='black', linestyle='--')
#ax2.axhline(y=0.89, color='black', linestyle='--')


plt.subplots_adjust(wspace=0.01)
plt.show()
```

```python
# ahora hago el grafico de potencia promedio
df = pd.DataFrame(WWZ_simple_linear[2])
df2 = pd.DataFrame(WWZ_simple_linear[1])
df11 = df.mean()
df22 = df2.mean()
time1 = df22.to_numpy() 
Averag = df11.to_numpy() 
```

```python
from scipy.signal import find_peaks

fig, ax = plt.subplots(figsize=(12, 4))

ax.plot(time1, Averag, 'ro', markersize=4)
ax.set_xlabel("Freq")
ax.set_ylabel('Power')
#ax.set_ylim([0, 10]) 

ax.axvline(x=0.12700, linewidth=2, color='black')
ax.axvline(x=0.222, linewidth=2, color='black')
ax.axvline(x=0.336, linewidth=2, color='black')
ax.axvline(x=0.422, linewidth=2, color='black')
ax.axvline(x=0.683, linewidth=2, color='black')

# Encontrar los picos
peaks, _ = find_peaks(Averag, height=0)

# Obtener las frecuencias y valores de potencia asociados a los picos
peaks_frequencies = [time1[peak] for peak in peaks]
peaks_powers = [Averag[peak] for peak in peaks]

# Agregar marcas con cruces en los picos
ax.scatter(peaks_frequencies, peaks_powers, marker='x', color='green', s=100)

# Imprimir los resultados de los picos
for freq, power in zip(peaks_frequencies, peaks_powers):
    print(f"Frecuencia: {freq}, Potencia: {power}")

plt.show()

# esto funciona perfectoo
```
